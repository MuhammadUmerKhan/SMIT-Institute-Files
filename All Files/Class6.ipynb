# Mean Median Mode

```python
data.fillna({'Region':data.Region.mode()[0], 
            'Manager':data.Manager.mode()[0],
            "SalesMan":data.SalesMan.mode()[0],
            "Item":data.Item.mode()[0],
            "Units":data.Units.mean(),
            "Unit_price":data.Unit_price.median(),
            "Sale_amt":data.Sale_amt.median(),
            "extra":300})
```
