```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
```

```python
data = pd.read_csv("./Ready for Model Building (Pakistan House DAtaset)")
data.head()
```

## Outliers

```python
for i in data.columns:
    if data[i].dtype != 'object':
        z_score = (data[i] - data[i].mean())/data[i].std()
        data['z_score'] = z_score
        df_clean = data[(data['z_score'] > -3) & (data['z_score'] < 3)]
df_clean.head()
```

```python
for i in df_clean.columns:
    if df_clean[i].dtype == 'object':
        le = LabelEncoder()
        df_clean[i] = le.fit_transform(df_clean[i])
df_clean.head()
```

```python
X = df_clean.drop(columns=['price', 'z_score'])
y = df_clean['price']
```

```python
scaler = StandardScaler()
X_transformed = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.33, random_state=42)
```

```python
def evaluate_model(X, y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
    models = {
        'LinearRegression': LinearRegression() , 'Ridge': Ridge() , 'Lasso': Lasso(), 
        'SVR': SVR(), 'RandomForestRegressor': RandomForestRegressor(), 'KNeighborsRegressor': KNeighborsRegressor()
    }
    result = []
    for model_name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)

        result.append({
            'Model': model_name,
            'R2_Score': r2_score(y_true=y_test, y_pred=y_pred),
            'Mean_Squared_Error': mean_squared_error(y_test, y_pred=y_pred)
        })
    result_df = pd.DataFrame(result)
    return result_df
```

```python
result_df = evaluate_model(X_transformed, y)
```
